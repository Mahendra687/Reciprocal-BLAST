**Creating a database from sequences** 

First you need to create a database. Therefore you change the directory into the dbdirectory where all the sequences are stored. Next you run the second command. This will create a database with all the downloaded sequences. If you downloaded nucleotide or protein sequence make sure you change the dbtype option as prot or nucl.

**Parameters**


1. **in:**FASTA file including all sequences for you database
2. **out:** Database name
3. **dbtype:** Database type usually 'prot' for proteins or 'nucl' for nucleotides

........................................................................................................
